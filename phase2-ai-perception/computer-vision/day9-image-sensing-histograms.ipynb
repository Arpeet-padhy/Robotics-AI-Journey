```python
# ============================================
# Phase 2 — Day 9: Image Sensing & Histograms
# Pixel intensity, noise, histograms, thresholding
# =============================================

import cv2 
import numpy as np 
import matplotlib.pyplot as plt

#Load our test image from Day 7 
img = cv2.imread('/home/arpeetpadhy/test_image.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print("Image loaded!")
print("Color Shape:", img.shape)
print("Grayscale shape:", gray.shape)
print("Pixel value range:", gray.min(),"To" ,gray.max())
```

```python
# ============================================
# IMAGE HISTOGRAMS
# ============================================

# A Histogram shows how many pixels have each intensity value (0-255)
hist = cv2.calcHist([gray], [0], None, [256],[0, 256])

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].imshow(gray, cmap='gray')
axes[0].set_title('Grayscale Image')
axes[0].axis('off')

#show the histogram 
axes[1].plot(hist, color= 'black')
axes[1].set_title('Pixel Intensity Histogram')
axes[1].set_xlabel('Pixel Intensity (0=black, 255= white)')
axes[1].set_ylabel('Number of Pixels')
axes[1].grid(True)

plt.tight_layout()
plt.show()

print("Histogram tells us the distridution of brightness in the image")
print("Peaks show common intensity values")
print(f"\nTotal pixels: {gray.shape[0] * gray.shape[1]}")
print(f"Sum of histogram: {int(hist.sum())}") # should match total pixels
```

```python
# ============================================
# THRESHOLDING — Binary Image Segmentation
# ============================================

# Simple threshold: pixels above value → white, below → black
_, thresh1 = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
_, thresh2 = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY)

fig, axes = plt.subplots(1, 3, figsize=(14,4))

axes[0].imshow(gray, cmap='gray')
axes[0].set_title('Original Grayscale')
axes[0].axis('off')

axes[1].imshow(thresh1, cmap='gray')
axes[1].set_title('Threshold = 100')
axes[1].axis('off')

axes[2].imshow(thresh2, cmap='gray')
axes[2].set_title('Threshold = 50')
axes[2].axis('off')

plt.suptitle('Thresholding — Separating Objects from Background', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("Thresholding converts grayscale to pure black/white")
print("Any pixel above the threshold value becomes white (255)")
print("Any pixel below becomes black (0)")
print("\nThis is the foundation of object segmentation in robotics vision")
```
